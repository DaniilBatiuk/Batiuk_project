1. У чому відмінність інтерпретатора від компілятора? Python – компілятор чи інтерпретатор?

- Інтерпретатор виконує код послідовно рядок за рядком в реальному часі. Він читає і виконує код без попередньої компіляції. Python - це інтерпретована мова програмування, що означає, що код Python виконується інтерпретатором Python.


- Компілятор перетворює вихідний код програми в машинний код (або інший мова програмування) попередньо, до запуску. Іншими словами, ви отримуєте виконуваний файл, який можна запустити без потреби в наявності вихідного коду. Python може використовувати компілятори, такі як Cython або Nuitka, для перетворення вихідного коду Python в бінарний код.

2. Що таке IDE PyCharm?

PyCharm - це інтегроване середовище розробки (IDE) для мови програмування Python, створене компанією JetBrains. Воно надає розширені засоби для розробки Python-програм, включаючи підсвічування синтаксису, автодоповнення, відладку, управління пакетами і інше. PyCharm доступний у двох версіях: Community (безкоштовна версія) і Professional (платна версія з додатковими функціями).

3. Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ SVN та GIT?

Система контролю версій (СКВ) - це інструмент для відстеження змін у вихідному коді або файловій структурі проекту з часом. 

- SVN (Subversion) - це централізована система контролю версій, де всі розробники взаємодіють з одним центральним репозиторієм. Зміни завантажуються в центральний репозиторій.


- Git - це розподілена система контролю версій, в якій кожен розробник має локальну копію всього репозиторію. Це дозволяє працювати офлайн та взаємодіяти з іншими локальними репозиторіями.

4. Що відбувається при виконанні команди git add .?

Команда git add . додає всі зміни і нові файли, які знаходяться у вашому поточному робочому каталозі, до стадії підготовки (staging area) в Git. Це означає, що зміни готові до включення в наступний коміт (збереження). Це дозволяє вам вибрати, які зміни ви хочете включити в наступний коміт.


5. Що відбувається при виконанні команди git commit?

Команда git commit зберігає всі зміни, які були додані до стадії підготовки (staging area), в історію репозиторію Git. Коміт створює новий "знімок" поточного стану робочого каталогу. Ви також повинні вказати повідомлення коміту, яке пояснює, які зміни ви зробили.

6. Якою Git-командою виконується розміщення локальної копії Git-репозиторію проєкта на віддаленому, що знаходиться на GitHub

Для розміщення локальної копії Git-репозиторію проєкта на віддаленому репозиторії GitHub використовується команда git push.
